```python
#The only library we need for converting an image into a pencil sketch with Python is an OpenCV library in Python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("nani.jfif")
cv2_imshow(image)
cv2.waitKey(0)
```

```python
#After reading the image, we will create a new image by converting the original image to greyscale
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

```python
#the next step is to invert the new grayscale image
inverted_image=255-gray_image
cv2_imshow(inverted_image)
cv2.waitKey(0)
```

```python
#The next step in the process is to blur the image by using the Gaussian Function in OpenCV
blurred=cv2.GaussianBlur(inverted_image,(21,21),0)
cv2_imshow(blurred)
cv2.waitKey(0)
```

```python
#Final step is to invert the blurred image, then we can easily convert the image into a pencil sketch
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(gray_image,inverted_blurred,scale=256.0)
```

```python
#FInal Result
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
